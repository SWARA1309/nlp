```python
# prompt: perform preprocessing for above 'Lowercase conversion• Removing HTML Tags. • Contraction mapping like conversion of haven’t to have not, We’re to we are, I’m to I am, he’s to he is, etc. •Removing “” i.e., the text having sentences written in direct speech. • Removing text inside the parenthesis. • Stop-Words and short words removal. • Punctuation Removal. • Removing special characters'

import re

def preprocess_text(text):
    # Lowercase the text
    text = text.lower()

    # Remove HTML tags
    text = re.sub('<.*?>', '', text)

    # Contraction mapping
    contractions = {
        "haven't": "have not",
        "We're": "We are",
        "I'm": "I am",
        "he's": "he is",
        "she's": "she is",
        "it's": "it is",
        "they're": "they are",
        "that's": "that is",
        "won't": "will not",
        "can't": "cannot",
        "n't": "not",
        "'s": " ",
        "'ve": "have",
        "'m": "am",
        "'re": "are",
    }
    for contraction, expansion in contractions.items():
        text = text.replace(contraction, expansion)

    # Remove text inside double quotes
    text = re.sub(r'"(.*?)"', '', text)

    # Remove text inside parentheses
    text = re.sub(r'\((.*?)\)', '', text)

    # Remove stop words and short words
    stop_words = set(stopwords.words('english'))
    short_words = set([word for word in stop_words if len(word) <= 2])
    text = ' '.join([word for word in text.split() if word not in (stop_words | short_words)])

    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # Remove special characters
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)

    return text

dataset_train_cnn['input_text'] = dataset_train_cnn['input_text'].apply(preprocess_text)
dataset_test_cnn['input_text'] = dataset_test_cnn['input_text'].apply(preprocess_text)
dataset_validation_cnn['input_text'] = dataset_validation_cnn['input_text'].apply(preprocess_text)
```

```python
# prompt: write code for performing title generation from article column

import numpy as np
def generate_title(article):
  # Preprocess the article text
  article_text = preprocess_text(article)

  # Tokenize the article text
  tokens = nltk.word_tokenize(article_text)

  # Get the word vectors for each token
  word_vectors = [model.wv[token] for token in tokens if token in model.wv.key_to_index]

  # Calculate the average word vector
  average_vector = np.mean(word_vectors, axis=0)

  # Find the closest word vector to the average vector
  closest_word = model.wv.similar_by_vector(average_vector, topn=1)[0][0]

  # Return the closest word as the title
  return closest_word

# Generate titles for the articles in the training set
titles = [generate_title(article) for article in dataset_train_cnn['article']]

# Add the generated titles to the dataset
dataset_train_cnn['title'] = titles

# Print the first 5 rows of the dataset
dataset_train_cnn.head()
```

```python
# prompt: The vector or the word embeddings so formed are passed into the encoder network i.e. has LSTM layers in it and provide an output feature vector that holds the necessary information to map the input to output.
import torch
import torch.nn as nn

# Define the encoder network
class Encoder(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers):
        super(Encoder, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)

    def forward(self, x):
        # Initialize the hidden and cell states
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)

        # Forward pass
        output, (hn, cn) = self.lstm(x, (h0, c0))

        # Return the output and the hidden state
        return output, hn

# Example usage
# Assuming word_embeddings is a tensor of shape (batch_size, sequence_length, input_size)
word_embeddings = torch.randn(10, 20, 300)  # Example random word embeddings
encoder = Encoder(input_size=300, hidden_size=256, num_layers=2)

# Pass the word embeddings through the encoder network
output, hn = encoder(word_embeddings)

# Print the shape of the output
print(output.shape)

```

```python
# prompt: The decoder uses the obtained representations from all the models i.e. it is able to access the hidden layers of the article. The next step is to convert the vector to words, so the decoder takes the words of the document summary and forms a context vector ct.

# Define the decoder network
class Decoder(nn.Module):
    def __init__(self, output_size, hidden_size, num_layers):
        super(Decoder, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(output_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x, encoder_outputs, hidden):
        # Forward pass
        output, hidden = self.lstm(x, hidden)
        prediction = self.fc(output)

        return prediction, hidden

# Example usage
# Assuming encoder_outputs is a tensor of shape (batch_size, sequence_length, hidden_size)
# and hidden is a tuple of (hn, cn) where hn and cn are tensors of shape (num_layers, batch_size, hidden_size)
encoder_outputs = torch.randn(10, 20, 256)  # Example random encoder outputs
hidden = (torch.randn(2, 10, 256), torch.randn(2, 10, 256))  # Example random hidden state
decoder = Decoder(output_size=300, hidden_size=256, num_layers=2)

# Pass the input through the decoder network
prediction, hidden = decoder(word_embeddings, encoder_outputs, hidden)

# Print the shape of the prediction
print(prediction.shape)
```

```python
# prompt: Next we have applied the attention-network after the encoder-decoder model.

import torch
import torch.nn as nn

# Define the attention network
class BahdanauAttention(nn.Module):
    def __init__(self, hidden_size):
        super(BahdanauAttention, self).__init__()
        self.hidden_size = hidden_size
        self.query_layer = nn.Linear(hidden_size, hidden_size)
        self.key_layer = nn.Linear(hidden_size, hidden_size)
        self.value_layer = nn.Linear(hidden_size, hidden_size)
        self.score_layer = nn.Linear(hidden_size, 1)

    def forward(self, query, keys, values):
        # Query: (batch_size, 1, hidden_size)
        # Keys: (batch_size, sequence_length, hidden_size)
        # Values: (batch_size, sequence_length, hidden_size)

        # Query layer
        query = self.query_layer(query)

        # Key layer
        keys = self.key_layer(keys)

        # Value layer
        values = self.value_layer(values)

        # Calculate scores
        scores = self.score_layer(torch.tanh(query + keys))

        # Apply softmax to normalize scores
        attention_weights = torch.softmax(scores, dim=1)

        # Calculate context vector
        context_vector = torch.sum(attention_weights * values, dim=1)

        return context_vector, attention_weights

# Define the decoder with attention
class DecoderWithAttention(nn.Module):
    def __init__(self, output_size, hidden_size, num_layers):
        super(DecoderWithAttention, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(output_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size * 2, output_size)
        self.attention = BahdanauAttention(hidden_size)

    def forward(self, x, encoder_outputs, hidden):
        # Forward pass through LSTM
        output, hidden = self.lstm(x, hidden)

        # Calculate attention context vector
        context_vector, attention_weights = self.attention(output.unsqueeze(1), encoder_outputs, encoder_outputs)

        # Expand output tensor to match the context_vector's shape
        output = output.expand(-1, context_vector.size(1), -1)

        # Concatenate context vector with LSTM output
        concat = torch.cat((context_vector, output), dim=-1)

        # Pass concatenated vector through a linear layer
        prediction = self.fc(concat)

        return prediction, hidden, attention_weights

# Example usage
# Assuming encoder_outputs is a tensor of shape (batch_size, sequence_length, hidden_size)
# and hidden is a tuple of (hn, cn) where hn and cn are tensors of shape (num_layers, batch_size, hidden_size)
encoder_outputs = torch.randn(10, 20, 256)  # Example random encoder outputs
hidden = (torch.randn(2, 10, 256), torch.randn(2, 10, 256))  # Example random hidden state
decoder = DecoderWithAttention(output_size=300, hidden_size=256, num_layers=2)

# Assuming word_embeddings is a tensor of shape (batch_size, sequence_length, output_size)
word_embeddings = torch.randn(10, 1, 300)  # Example random word embeddings

# Pass the input through the decoder with attention
prediction, hidden, attention_weights = decoder(word_embeddings, encoder_outputs, hidden)

# Print the shape of the prediction
print(prediction.shape)

```
